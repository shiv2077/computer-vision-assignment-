```python
# Import YOLO from ultralytics
from ultralytics import YOLO

# Path to the YAML configuration file
data_yaml = "/content/drive/MyDrive/medicinal plant classification /data.yaml"  # Update this path to your YAML file

# Step 1: Load the YOLO model
model = YOLO("yolo11n.pt")  # You can use yolov8s.pt or yolov8x.pt for better accuracy with more compute power

# Step 2: Train the model
model.train(
    data=data_yaml,  # Path to the YAML file
    epochs=50,       # Number of training epochs
    imgsz=640,       # Image size
    batch=16,        # Batch size (adjust based on your GPU memory)
    device=0,        # Use GPU if available
    workers=4,       # Number of CPU workers
    name='yolov11_medicinal_plants',  # Experiment name
    save_period=1,   # Save checkpoint every epoch
    augment=True     # Enable data augmentation
)

# Step 3: Evaluate the model on the validation set
model.val()

# Step 4: Run inference on a test image
test_image_path = "/content/drive/MyDrive/medicinal plant classification /valid/images/1010_jpg.rf.7c7f2d5254bf36225c42d126f8ca5dca.jpg"  # Replace with your test image path
results = model.predict(test_image_path)

# Display the prediction results
for result in results:  # Iterate over the list of results
    result.show()  # Call show() on each result object


# Optional: Save the trained model to Google Drive
model.save("/content/drive/MyDrive/yolov11_medicinal_plants_best.pt")

# Step 5: Plot training results (accuracy, loss, etc.)
import matplotlib.pyplot as plt
history = model.history  # Training history

# Plot training and validation losses
if history:
    plt.plot(history['train_loss'], label='Train Loss')
    plt.plot(history['val_loss'], label='Validation Loss')
    plt.legend()
    plt.title("Training Loss vs Validation Loss")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.show()
```
